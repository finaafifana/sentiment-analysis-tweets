```python
#search_key = api.user_timeline(id="america",count=10)
search_key = api.search(q="sarcasm",lang="en", count=100)
search_key
```

```python
import re,string

#create variabel global as list for remove duplicate tweets
analysis_result = []

for tweet in search_key:
    tweet_properties = {} #split tweets to dict as dictionary
    tweet_properties["date"] = tweet.created_at
    tweet_properties["username"] = tweet.user.screen_name
    tweet_properties["teks"] = tweet.text
    #print(tweet_properties) 
    #print(tweet.user.screen_name, "NGETWEET",tweet.text) show tweets by username : ngetweet
    
    #if you print(tweet.text) data is still dirty, so we clean data first to remove unusefull characters like emojis,@,hashtag,ect
    tweet_clean = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet.text).split())
    #print(tweet_clean)
    
    analysis = TextBlob(tweet_clean) #determine the polarity of tweets
    #print(analysis.sentiment.polarity)
    
    if analysis.sentiment.polarity>0.0:
        tweet_properties["sentiment"] = "positive"
    elif analysis.sentiment.polarity==0.0:
        tweet_properties["sentiment"] = "netral"
    elif analysis.sentiment.polarity<0.0:
        tweet_properties["sentiment"] = "negative"
    
    #remove duplicate tweet such as RT
    if tweet.retweet_count > 0:
        if tweet_properties not in analysis_result:
            analysis_result.append(tweet_properties)
    else:
        analysis_result.append(tweet_properties)
```
